# Understanding Time Complexity with Bubble Sort

In this notebook, we will use Bubble Sort to understand how time complexity is related to input size.

We will calculate the time taken by the algorithm for different input sizes and derive an equation for the time complexity. We will also explore why lower-order terms don't matter.

```python
import time
import numpy as np
import matplotlib.pyplot as plt

def bubble_sort(arr):
    """Bubble Sort algorithm to sort an array."""
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def measure_time_bubble_sort(input_sizes):
    """Measure the time taken for Bubble Sort for different input sizes."""
    times = []
    for size in input_sizes:
        arr = np.arange(size, 0, -1)  # Create an array from size to 1
        start_time = time.time_ns()  # Get the start time in nanoseconds
        bubble_sort(arr)  # Sort the array using Bubble Sort
        end_time = time.time_ns()  # Get the end time in nanoseconds
        times.append(end_time - start_time)
    return times

# Define input sizes (small for quick demonstration)
input_sizes = [10, 100, 500, 1000, 2000,3000,10000]

# Measure the time taken for each input size
times = measure_time_bubble_sort(input_sizes)

# Plot the results
plt.plot(input_sizes, times, marker='o')
plt.xlabel('Input Size (n)')
plt.ylabel('Time Taken (nanoseconds)')
plt.title('Time Complexity of Bubble Sort')
plt.grid(True)
plt.show()
```

### Analyzing the Time Complexity Equation

Now let's derive an equation for the time complexity of Bubble Sort. 

1. **Bubble Sort Time Complexity**: 
   - Bubble Sort works by comparing adjacent elements in the array and swapping them if they are in the wrong order.
   - In the worst case, Bubble Sort has to perform `n-1` comparisons in the first iteration, `n-2` in the second iteration, and so on.
   - This gives us a total of `n * (n-1) / 2` comparisons.
   - Therefore, the time complexity is O(n^2).

2. **Why Lower-Order Terms Don't Matter**: 
   - In Big-O notation, we focus on the term that grows the fastest as the input size increases.
   - For Bubble Sort, the dominant term is `n^2`. Lower-order terms (like `n` or constants) become insignificant as `n` grows larger.
   - For example, if `n = 1000`, the difference between `n^2` (which is 1,000,000) and `n` (which is just 1000) is huge. The lower-order term (`n`) has little impact on the overall complexity.

### Demonstrating the Impact of Lower-Order Terms

To further illustrate this point, let's plot the equation for `n^2 + n` and see how the `n^2` term dominates as `n` grows.

```python
# Demonstrating the impact of lower-order terms
n_values = np.arange(1, 400)
n_squared = n_values ** 2
n_plus_n_squared = n_values ** 2 + n_values

# Plotting the results
plt.plot(n_values, n_squared, label='n^2')
plt.plot(n_values, n_plus_n_squared, label='n^2 + n', linestyle='--')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.title('Impact of Lower-Order Terms')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Demonstrating the impact of lower-order terms
n_values = np.arange(1, 100000)
n_squared = n_values ** 2
n_plus_n_squared = n_values ** 2 + n_values*100

# Plotting the results
plt.plot(n_values, n_squared, label='n^2')
plt.plot(n_values, n_plus_n_squared, label='n^2 + 100*n', linestyle='--')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.title('Impact of Lower-Order Terms')
plt.legend()
plt.grid(True)
plt.show()
```

As you can see from the graph, as `n` grows larger, the `n^2` term (solid line) quickly outgrows the `n` term. This is why we only focus on the highest-order term when analyzing time complexity.
